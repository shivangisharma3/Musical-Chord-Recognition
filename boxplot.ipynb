```python
import numpy as np
import matplotlib as mpl
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

import json

chords=['c','c#','d','d#','e','f','f#','g','g#','a','a#','b','cm','c#m','dm','d#m','em','fm','f#m','gm','g#m','am','a#m','bm']
notes=['C','C#','D','D#','E','F','F#','G','G#','A','A#','B']
starthop=512
for i in range(len(chords)):
    with open("guitar_clean_"+str(starthop)+"/"+chords[i]+".json") as readfile :
        a = readfile.read()
    a = a.strip()
    x = json.loads(a)
    plt.figure(num=i, figsize=(8,8))
    df = pd.DataFrame(x)
    df.columns = ['C','C#','D','D#','E','F','F#','G','G#','A','A#','B']
    mid=(i+4)%12
    cname=notes[i%12]
    if i > 11:
        mid=(3+i)%12
        cname=cname+" Minor"
    else:
        cname=cname+" Major"
    titl=[cname, "(",notes[i%12],"-",notes[mid],"-",notes[(i+7)%12],")"]
    titl=" ".join(titl)
    ax = sns.boxplot( data = df,orient = 'v').set_title(str(titl))
    
```
